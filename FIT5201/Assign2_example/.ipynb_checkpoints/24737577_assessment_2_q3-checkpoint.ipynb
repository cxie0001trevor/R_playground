__Question 3.2__ 

Train an autoencoder (similar to Activity 5.2) with only one hidden layer and change the number of its neurons to: 20, 40, 60, 80, .., 500 (i.e. from 20 to 500 with a step size of 20). (see 3.3)

__Question 3.3__ 

For each model in Step 3.2, calculate and record the reconstruction error which is simply the average (over all data points while the model is fixed) of Euclidian distances between the input and output of the autoencoder (you can simply use h2o.anomaly() function). Plot these values where the x-axis is the number of units in the middle layer and the y-axis is the reconstruction error. Then, save and attach the plot to your Notebook file, and explain your findings based on the plot.

```r
r.error <- matrix(NA, nrow=length(seq(20,500,20)), ncol=1)  # create an empty list to store all models 

for (k in seq(20, 500, 20)){                                # for different neurons in the hidden layer seq(20,500,20)
    
    # 3.2 train the model
    model <- h2o.deeplearning(                              # build a NN classifier based on the labeled training data
        x = 1:ncol(train.data),                             # specify the dataset for training 
        training_frame = train.data,                        # specify the frame (imported file)    
        hidden = c(k),                                      # number of layers and their units, here only 1 layer
        epochs = 50,                                        # maximum number of epoches  
        activation = 'Tanh',                                # activation function 
        autoencoder = TRUE,                                 # use autoencoder
        l2 = 0.1                                            # specify the L2 regularisation weight to add stability    
    )
    
    # 3.3 claculate the mean of the reconstruction errors
    r.error[k/20] <- mean(h2o.anomaly(model, train.data, per_feature = FALSE)) 
}
```

```r
plot(seq(20, 500, 20), r.error, type='l', xlab = 'Number of neurons in hidden layer', ylab = 'Average reconstruction error')
```

<font color='blue'>Answer: </font> the graph shows a raid drop at the beginning followed by reaching a global minimum at hidden = 120. The average reconstruction error at this point is roughly 0.005. The second loweres point is at 200, i.e. the 10th iteration and the error at this point is really close to that of the global minimum. Then the graph gradually climbs up to around 0.01. Hence the best number of hidden units in the hidden layer is 120. 

__Question 3.4__ 

Use the 3-layer NN from Activity 3.1 or h2o.deeplearning function (make sure you set  autoencoder = FALSE) to build a model with 100 units in the hidden layer using all the original attributes from the training set. Then, calculate and record the test error.

```r
# build a neural network classifier based on the labeled training data
NN.model <- h2o.deeplearning(    
  x = 2:ncol(train.data),           # specify training data without the label
  y = 1,                            # specify training label
  training_frame = train.data,      # specify the frame    
  hidden = c(100),                  # number of layers and their units, here only 1 layer
  epochs = 50,                      # maximum number of epoches  
  activation = 'Tanh',              # activation function 
  autoencoder = FALSE,              # autoencoder is FALSE
  l2 = 0.1                          # specify the L2 regularisation weight to add stability
)
```

```r
# the error function
error.rate <- function(Y1, T1){
  if (nrow(Y1)!=nrow(T1)){
    stop('error.rate: size of true lables and predicted labels mismatch')
  }
  return (sum(T1!=Y1)/nrow(T1))
}
```

```r
# make a prediction on the test set
test.predict <- h2o.predict(NN.model, test.data)$predict
# calculate the error rate
test.error <- error.rate(test.data$label, test.predict)
```

__Question 3.5__ 

Build augmented self-taught networks using the models learnt in Step 3.2. For each model:
* Add the output of the middle layer as extra features to the original feature set,
* Train a 3-layer NN (similar to Step 3.4) using all features (original + extra). Then calculate and record the test error.

```r
# normalise the dataset 
train.data[,-1] <- scale(train.data[,-1])
test.data[,-1] <- scale(test.data[,-1])
```

```r
# the error function
error.rate <- function(Y1, T1){
  if (nrow(Y1)!=nrow(T1)){
    stop('error.rate: size of true lables and predicted labels mismatch')
  }
  return (sum(T1!=Y1)/nrow(T1))
}
```

```r
tt.error <- matrix(NA, nrow=length(seq(20,500,20)), ncol=2) # create an empty list to store all models 

for (k in seq(20, 500, 20)){                                # for different neurons in the hidden layer seq(20,500,20)
    
    # train the model
    model <- h2o.deeplearning(                              # build a NN classifier based on the labeled training data
        x = 1:ncol(train.data),                             # specify the dataset for training 
        training_frame = train.data,                        # specify the frame (imported file)    
        hidden = c(k),                                      # number of layers and their units, here only 1 layer
        epochs = 50,                                        # maximum number of epoches  
        activation = 'Tanh',                                # activation function 
        autoencoder = TRUE,                                 # use autoencoder
        l2 = 0.1                                            # specify the L2 regularisation weight to add stability    
    )
    
    # extract the middle layer as extra features
    feature <- h2o.deepfeatures(model, train.data, layer=1)

    
    # combine the feature with original data
    new.train.data <- h2o.cbind(train.data, feature)        # append all together
    tt.error[k/20, 1] <- dim(new.train.data)[2] - 1         # record the number of features in the training set
    
    # train the model with extra features
    new.model <- h2o.deeplearning(                          # build a NN classifier based on the labeled training data
        x = 2:ncol(new.train.data),                         # specify the dataset for training 
        y = 1,                                              # specify training label
        training_frame = new.train.data,                    # specify the frame (imported file)    
        hidden = c(k),                                      # number of layers and their units, here only 1 layer
        epochs = 50,                                        # maximum number of epoches  
        activation = 'Tanh',                                # activation function 
        autoencoder = FALSE,                                # autoencoder is False
        l2 = 0.1                                            # specify the L2 regularisation weight to add stability    
    )
    
    # calculate the test error
    test.predict <- h2o.predict(new.model, test.data)$predict
    tt.error[k/20, 2] <- error.rate(test.data$label, test.predict)
    
}
```
