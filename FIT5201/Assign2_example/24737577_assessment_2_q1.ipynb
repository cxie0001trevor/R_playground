First, let's define some auxilary functions.

```r
# Auxilary function that is the objective function for the EM algorithm
train_obj <- function(model, counts) { 
  N <- dim(counts)[2] # number of documents
  K <- dim(model$mu)[1]
   
  nloglike = 0
  for (n in 1:N){
    lprob <- matrix(0,ncol = 1, nrow=K)
    for (k in 1:K){
      lprob[k,1] = sum(counts[,n] * log(model$mu[k,])) 
    }
    nloglike <- nloglike - logSum(lprob + log(model$rho))
  }
  
  return (nloglike)
}
```

```r
# Auxilary function to prevent numerical overflow/underflow when working with small numbers
logSum <- function(v){
   m = max(v)
   return (
       m + log(sum(exp(v-m))))
}
```

```r
# Auxilary function that initialises the parameters for the algorithm.
initial.param <- function(vocab_size, K){
  rho <- matrix(1/K,nrow = K, ncol=1)                              # assume all clusters have the same size 
  mu <- matrix(runif(K*vocab_size),nrow = K, ncol = vocab_size)    # initiate Mu 
  mu <- prop.table(mu, margin = 1)                                 # normalization to ensure that sum of each row is 1
  return (list("rho" = rho, "mu"= mu))                             # mu is randomly selected class mean
}
```

Now let's write the core EM functions.

```r
softE.step <- function(gamma, model, counts){         # this function currently implements the E-step of the soft-EM
  N <- dim(counts)[2]                                 # number of documents
  K <- dim(model$mu)[1]                               # number of dimensions 

  for (n in 1:N){
    for (k in 1:K){
      ## calculate the posterior based on the estimated mu and rho in the "log space"
      gamma[n,k] <- log(model$rho[k,1]) +  sum(counts[,n] * log(model$mu[k,])) 
    }
    
    logZ = logSum(gamma[n,])                          # normalisation to sum to 1 in the log space
    gamma[n,] = gamma[n,] - logZ
  }
  
  gamma <- exp(gamma)                                 # converting back from the log space 
    
  return (gamma)
}
```

```r
hardE.step <- function(gamma, model, counts){         # this function implements the E-step of the hard-EM
  N <- dim(counts)[2]                                 # number of documents
  K <- dim(model$mu)[1]                               # number of dimensions 
  
  for (n in 1:N){
      for (k in 1:K){
      # calculate the posterior based on the estimated mu and rho in the "log space"
      gamma[n,k] <- log(model$rho[k,1]) +  sum(counts[,n] * log(model$mu[k,])) 
      }
      
      logZ = logSum(gamma[n,])                        # normalisation to sum to 1 in the log space
      gamma[n,] = gamma[n,] - logZ
  }
  
  gamma <- exp(gamma)                                 # converting back from the log space 
    
  # for the hard EM, we assign the maximum probability to be 1 (here something close to 1 to avoid mathmetical error)
  max.prob <- gamma==apply(gamma, 1, max)             # for the class label with the maximum (estimated) probability
  gamma[max.prob] <- 0.999997                         # re-assign its probability to 1
  gamma[!max.prob] <- 0.000001                        # re-assign the others to 0
    
  return (gamma)
}
```

```r
M.step <- function(gamma, model, counts){
    # update p(w|z)
    N <- dim(counts)[2]                                                            # number of documents
    W <- dim(counts)[1]                                                            # number of words i.e. vocabulary size
    K <- dim(model$mu)[1]                                                          # number of clusters
                                     
    for (k in 1:K){
        ## recalculate the estimations:
        model$rho[k] <- sum(gamma[,k])/N                                            # the relative cluster size                        
        model$mu[k,] <- (gamma[,k] %*%  t(counts)) / sum(gamma[,k] %*%  t(counts))  # new means (cluster cenroids)
        }
    model$mu <- scale(model$mu+1e-10, center = F, scale=colSums(model$mu+1e-10))    # normalise the mu
    
  
  return (model) # Return the new model
}
```

Now, let's define another function that wraps everything we have written together.

```r
EM <- function(counts, K, type, max.epoch=50, epsilon = 0.01, random.state=NULL, print_error=F){
                         
    # Model Parameter Setting
    N <- dim(counts)[2]                                                    # number of documents
    W <- dim(counts)[1]                                                    # number of unique words (in all documents)
    eta <- 1                                                               # epoch counter
    terminate <- FALSE                                                     # set terminate condition
    model <- initial.param(W, K)                                           # Initialise the model
    gamma <- matrix(0, nrow = N, ncol = K)                                 # Initialise the gamma
    set.seed(random.state)                                                 # set the random state if given
    
    
    # Build the model
    while(!terminate){
        
        # E Step
        if (type == 'soft'){
            gamma <- softE.step(gamma, model, counts) 
        } else {
            gamma <- hardE.step(gamma, model, counts) 
        }
        
        old.mu <- model$mu                                                 # recode the old mu for epsilon comparison
        
        # M Step
        model <- M.step(gamma, model, counts)
        
        if (print_error){                                                   # print the error as specified
            print(train_obj(model,counts))
        }
        
        terminate <- eta>max.epoch | sum(abs(old.mu - model$mu))<=epsilon   # update the terminate condition
        
        eta <- eta + 1                                                      # update the counter
      }
    
  cat('The maximum number of iterations is:', eta,'\n')  
  
  return(list("model"=model,"gamma"=gamma))                                 # return the new model and the old gamma
}
```

__Reading the File and Feature Extraction__: Prior doing any clustering, we need to extract some meaningful features from the datasets. The features should be numerical, otherwise our Euclidean distance (non-similarity) measure will not work. To do so, we use a popular R package called `tm` to tailor the texts and calculate word counts (frequency of each word in each single document) as the features.

```r
# Auxilary function that reads and preprocesses a text file 
readfile <- function(path, spr.ratio = 0.9){                   # Read and preprocess a text file 
    library(tm)                                                 # load Text Mining library, also requires "snowballc" library

    text <- readLines(path)
    docs <- strsplit(text, '\t')                                # the terms before '\t' are the lables
    labels <-  unlist(lapply(docs, function(x) x[1]))           # all the remaining text after '\t' are the actual documents
    docs <- data.frame(unlist(lapply(docs, function(x) x[2])))
    rm(text)                                                    # just free some memory!
    
    docs <- DataframeSource(docs)                               # create a corpus
    docs <- Corpus(docs)                                     

    # Preprocessing:
    docs <- tm_map(docs, removeWords, stopwords("english"))     # remove stop words
    docs <- tm_map(docs, removePunctuation)                     # remove pnctuation
    docs <- tm_map(docs, stemDocument)                          # perform stemming 
    docs <- tm_map(docs, removeNumbers)                         # remove all numbers
    docs <- tm_map(docs, stripWhitespace)                       # remove redundant spaces 

    # Create a documetn-term matrix (dtm) which its rows are the documents and colomns are the words. 
    dtm <- DocumentTermMatrix(docs)

    dtm <- removeSparseTerms(dtm, spr.ratio)                    # the threshold of relative document frequency for a term
    
    # convert the dtm to a term frequency-inverse document frequency (tf-idf) matrix
    # the rows are the words (129) and columns are the documents (2373)
    m <- t(as.matrix(dtm))
       
    return (list("tfidf"= m, "labels" = labels))                               
}
```
