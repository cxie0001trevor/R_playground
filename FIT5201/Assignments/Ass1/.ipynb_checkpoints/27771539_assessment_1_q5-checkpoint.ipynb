# Section D: Ridge Regression

## 1. The Gradient Method For Regularised Linear Regression

As for linear regression, the task of either minimizing or maximizing the objective function $E(\pmb{w})$ by altering $\pmb{w}$ is so called optimization problem. By using gradient method, we can find the optimal solution by updating $\pmb{w}$ iteratively, instead of generalising some formulae to solve it numerically.  
In this task, we use $\ell{2}$ regulariser to control our model complexity.

### 1.1 The  Batch Gradient Descent (BGD)

In BGD algorithm, each update is based on evaluating the entire training set. Due to the technique that use the whole data set at once, it is known as 'batch' method.  
The error function is definded with repect to the entire training set:

$$E(\pmb{w})=\sum_{i=1}^{N}{E_{i}(\pmb{w})}$$
Hence, the problem of fitting a model becomes:
$$\pmb{w}^{Ridge}=\underset{\pmb{w}\in \mathbb{R}}{argmin}\left\{E(\pmb{w})\right\}$$

At each iteration (step), $\pmb{w}$ got updated in the direction of the negative gradient of the objective function at current $\pmb{w}$:
$$\pmb{w}^{(\tau+1)} = \pmb{w}^{(\tau)}-\eta\nabla{E(\pmb{w}^{(\tau)})}$$
To avoid overfitting, the penalty term is included in the definition of the error function (the coefficient $\frac{1}{2}$ is for latter convenience), Then, we can have:
$$E(\pmb{w})=\frac{1}{2}\frac{1}{N}\sum_{i=1}^{N}{E_{i}(\pmb{w})}+penalty=\frac{1}{2}\frac{1}{N}\sum_{i=1}^{N}{\left[t_{i}-\pmb{w}\pmb{\phi}_{i}\right]^{2}}+\frac{\lambda}{2}\pmb{w}^{T}\pmb{w}$$

$\lambda$ is a complexity parameter that controls the amount of parameter, preventing from picking arbitrarily large parameter values. Hence, $\lambda$ controls the amount of the value shrinkage in parameters: larger the value of $\lambda$, the greater the amount of shrinkage.

Notice that the error function is differentiable, hence the derivative respect to $\pmb{w}$ :
$$\nabla{E(\pmb{w})}=\frac{\partial}{\partial\pmb{w} } E(\pmb{w})=-\frac{1}{N}\sum_{i=1}^{N}{\left[t_{i}-\pmb{w}\pmb{\phi}_{i}\right]}\pmb{\phi}_{i}+\lambda\pmb{w}$$

To wrap up, at each step of update, by the use of the derivative we have:
$$\pmb{w}^{(\tau+1)} = \pmb{w}^{(\tau)}-\eta\left[-\frac{1}{N}\sum_{i=1}^{N}{\{t_{i}-\pmb{w}^{(\tau)}\pmb{\phi}_{i}\}}\pmb{\phi}_{i}+\lambda\pmb{w}^{(\tau)} \right]$$

This simultaneous update is one of the key property of gradient descent. Notice that $\pmb{w}$ is a vector of numbers of parameters $w_{i}$. All parameters should be updated simultaneously at each step.  (*Quote from [Stanford University-Linear Regression-Andrew Ng](https://www.youtube.com/watch?v=W46UTQ_JDPk)*)

### 1.2 The Stochastic Gradient Descent (SGD)

Compared with BGD, SGD makes update to $\pmb{w}$ based on one data point at a time:
$$\pmb{w}^{(\tau+1)} = \pmb{w}^{(\tau)}-\eta\nabla{E_{i}(\pmb{w}^{(\tau)})}$$
Then, the error function revise into:
$$E_{i}(\pmb{w})=\frac{1}{2}(t_{i}-\pmb{w}\pmb{x}_{i})^{2}+\frac{\lambda}{2}\pmb{w}^{T}\pmb{w} $$

the derivative respect to $\pmb{w}$ :
$$\nabla{E_{i}(\pmb{w})}=\frac{\partial}{\partial\pmb{w} } E_{i}(\pmb{w})=-(t_{i}-\pmb{w}\pmb{x}_{i})\pmb{x}_{i}+\lambda\pmb{w}$$
To wrap up, at each step of update, by the use of the derivative we have:
$$\pmb{w}^{(\tau+1)} = \pmb{w}^{(\tau)}-\eta\left[-(t_{i}-\pmb{w}^{(\tau)}\pmb{x}_{i})\pmb{x}_{i}+\lambda\pmb{w}^{(\tau)}\right]$$

## 2. Implement BGD and SGD in R

### 2.0 Auxilary Functions

```r
predict_func <- function(Phi, w){
    # For both SGD and BGD
    return(Phi %*% w)
} 

obj_func <- function(Phi, w, label, lambda){
    # L2 regulariser applied
    return(0.5 * mean((predict_func(Phi,w)-label)^2) + 0.5 * lambda * (w %*% w))
}

get_errors <- function(Phi, train.label, Phi_ts, test.label, W){
    # For Both
    n_w= nrow(W)
    errors = data.frame(matrix(NA, nrow=n_w, ncol=3))
    colnames(errors) <- c('tau', 'train', 'test')
    for (tau in 1:n_w) {
        errors[tau,1] = tau
        errors[tau,2] = obj_func(Phi, W[tau,], train.label, 0)
        errors[tau,3] = obj_func(Phi_ts, W[tau,], test.label, 0)
    }
    # return df of error for both train and test
    return(errors)
}

updatew_func <- function(w, Phi, label, t_pred, lambda, eta){
    # For BGD
    gradient = -colMeans(matrix((label-t_pred), nrow(Phi), ncol(Phi)) * Phi)
    # Get w updated: L2 regulariser applied
    new_w <- w - eta * (gradient + lambda * w)
    return(new_w)
}
```

### 2.1 BGD Algorithm

```r
BGD_reg <- function(train.data, train.label, test.data, test.label, lambda, eta, tau.max, isprint=T){
    s <- proc.time()
    N <- nrow(train.data)
    epsilon <- 0.001 * mean(train.label) 
    
    Phi <- as.matrix(cbind('x0'=1, train.data))
    Phi_ts <- as.matrix(cbind('x0'=1, test.data))
    
    # Initialise weights
    W <- matrix(NA,nrow=tau.max, ncol=ncol(Phi))
    #Using same starting weight for comparison
    W[1,] <- rep(0.001, ncol(Phi))
    
    for(tau in 1:tau.max){
        # check termination criteria:
        if (tau < tau.max){
            if(obj_func(Phi, W[tau,], train.label, lambda) <= epsilon){
                #print(paste('• Reach Theshold: cost <', epsilon))
                break
            }
            # At each iteration it evaluate the entire dataset
            t_pred <- predict_func(Phi, W[tau,])
            # Update w for next iteration
            W[tau+1,] <- updatew_func(W[tau,], Phi, train.label, t_pred, lambda, eta)
            # Stop at (tau.max-1)th iteration
        } else{
            #print(paste('• Reach Maximum tau',tau))
            break
        }
    }
    e <- proc.time() - s
    # Get df of error
    errors <- get_errors(Phi, train.label, Phi_ts, test.label, W)
    # Print find result in text
    opt_tau <- which.min(errors[,'test']); w <- W[opt_tau,]
    cat('\n-----------------------------BGD--------------------------------\n')
    cat('Finish in ', round(e[[3]],6), ' sec.\n')
    cat('Learning rate:', eta, 'Optimal Found @', opt_tau, '\nThe  final coefficents are:', w, 
        '\nTrain error: ', errors[opt_tau, 'train'], '\nTest error: ', errors[opt_tau, 'test'])
    
    
    # Plot the result
    error.m <- melt(errors, id='tau') # Manipulate df for plotting
    options(repr.plot.width=10, repr.plot.height=6, warn=-1) # More control on output
    # Plot line and the optimal solution
    p <- ggplot(data=error.m, aes(x=tau, y=value, color=variable)) +
            geom_line() + ggtitle(paste('BGD - Test and Train Error @learning rate=',eta)) +
            scale_color_discrete(guide = guide_legend(title = 'Errors')) +
            geom_point(x=opt_tau, y=errors[opt_tau, 'test'], color='red',size=3) + 
            theme_minimal()
    # To plot
    if(isprint){print(p)}
    # return the df
    
    return(errors)
}
```

### 2.2 SGD Algorithm

```r
SGD_reg <- function(train.data, train.label, test.data, test.label, lambda, eta, tau.max, isprint=T){
    s <- proc.time()
    N <- nrow(train.data)
    epsilon <- 0.001 * mean(train.label) 
    
    index <- sample(N, tau.max,replace=F)
    Phi <- as.matrix(cbind('x0'=1, train.data))
    Phi_ts <- as.matrix(cbind('x0'=1, test.data))
    
    # Initialise weights
    W <- matrix(NA,nrow=tau.max, ncol=ncol(Phi))
    #Using same starting weight for comparison
    W[1,] <- rep(0.001, ncol(Phi))
    
    # For record error values
    error <- data.frame('tau'=1:tau.max)
    
    Phi_ <- Phi[index,]
    label_ <- train.label[index]
    for(tau in 1:tau.max){
        # check termination criteria:
        if (tau < tau.max){
            if(obj_func(Phi, W[tau,], train.label, 0) <= epsilon){
                #print(paste('• Reach Theshold: cost <', epsilon))
                break
            }
            # At each iteration it evaluate the entire dataset
            t_pred <- predict_func(Phi_[tau,], W[tau,])
            # Update w for next iteration
            gradient = -as.vector(label_[tau] - t_pred) * Phi_[tau,]
            # Get w updated: L2 regulariser applied
            W[tau+1,] <- W[tau,] - eta * (gradient + lambda * W[tau,])
            # Stop at (tau.max-1)th iteration
        } else{
            #print(paste('• Reach Maximum tau',tau))
            break
        }
    }
    e <- proc.time() - s
    errors <- get_errors(Phi, train.label, Phi_ts, test.label, W)
    # Print find result in text
    opt_tau <- which.min(errors[,'test']); w <- W[opt_tau,]
    cat('\n-----------------------------SGD--------------------------------\n')
    cat('Finish in ', round(e[[3]],6), ' sec.\n')
    cat('Learning rate:', eta, 'Optimal Found @', opt_tau, '\nThe  final coefficents are:', w, 
        '\nTrain error: ', errors[opt_tau, 'train'], '\nTest error: ', errors[opt_tau, 'test'])
    
    # Plot the result
    error.m <- melt(errors, id='tau') # Manipulate df for plotting
    options(repr.plot.width=10, repr.plot.height=6, warn=-1) # More control on output
    # Plot line and the optimal solution
    p <- ggplot(data=error.m, aes(x=tau, y=value, color=variable)) +
            geom_line() + ggtitle(paste('SGD - Test and Train Error @learning rate=',eta)) +
            scale_color_discrete(guide = guide_legend(title = 'Errors')) +
            geom_point(x=opt_tau, y=errors[opt_tau, 'test'], color='red', size=3) + 
            theme_minimal()
    # To plot
    if(isprint){print(p)}
    # return the df
    return(errors)
}    
```
