## 2. Training autoencoder with different number of neurons

```r
# initialise the package
localH2O = h2o.init(nthreads = -1, ip = 'localhost', port = 54321)
```

```r
# load csv and remove NAs
labeled <- na.omit(read.csv('./datasets/Task2C_labeled.csv'))
unlabeled <- na.omit(read.csv('./datasets/Task2C_unlabeled.csv'))
test <- na.omit(read.csv('./datasets/Task2C_test.csv'))

# convert to h2o object
labeled <- as.h2o(labeled)     
unlabeled <- as.h2o(unlabeled)
test <- as.h2o(test)
```

```r
# Preparing data
train <- h2o.rbind(labeled, unlabeled) 
# convert label as factor
train$label <- as.factor(train$label)
test$label <- as.factor(test$label)
```

```r
# initialise a matrix for storing error of all models
error <- matrix(NA, nrow=25, ncol=1)

# a list for storing model for later purpose
model.list <- list()
# training ...
for (i in 1:25){
    # different neurons in the hidden layer
    # NN classifier
    model <- h2o.deeplearning(
        x = 1:ncol(train), # label and all pixels 
        training_frame = train, # specify the frame (imported file)    
        hidden = c(20*i), # number of layers and their units
        epochs = 50, # maximum number of epoches  
        activation = 'Tanh', # activation function 
        autoencoder = TRUE, # use autoencoder   
    )
    # store it for later use
    model.list[[i]] <- model
    
    # mean of the reconstruction errors
    error[i] <- mean(h2o.anomaly(model, train, per_feature = FALSE)) 
}
```

### Visualising

```r
# plot the error
error.m <- data.frame(k=seq(20,500,20), e=error)
# best k and its error
best_k <- data.frame(k=20*which.min(error),e=min(error))
ggplot(data=error.m, aes(x=k,y=e)) + geom_line() + 
    ggtitle('The Reconstruction error with various neurons in hidden layer') +
    labs(x='Number of neurons(K)', y='Average Error')+
    geom_point(color='blue') + 
    geom_point(aes(x=best_k$k, y=best_k$e), size = 10, colour="red", shape=1) +
    geom_label(aes(x=0.4*best_k$k, y=best_k$e), 
               label = paste("Best k: ",best_k$k, '\nError: ', round(100 * best_k$e,2), '%')) +
    theme_minimal()
```

## 3. 3-layers Neural Network with 100 units in hidden layer

```r
# the error function
error.rate <- function(Y1, T1){
    if (nrow(Y1)!=nrow(T1)){
    stop('error.rate: size of true lables and predicted labels mismatch')
    }
    return (sum(T1!=Y1)/nrow(T1))
}
```

```r
NN.model <- h2o.deeplearning(
    x = 2:ncol(train), # all pixels without label
    y=1, #training label
    training_frame = train, # specify the frame (imported file)    
    hidden = c(100), # number of layers and their units, here only 1 layer
    epochs = 50, # maximum number of epoches  
    activation = 'Tanh', # activation function 
    autoencoder = FALSE, # use autoencoder  
)
```

```r
# Use NN.model to predict test data
pred <- h2o.predict(NN.model, test)$predict
# error rate
test.error <- error.rate(test$label, pred)
# print the error(%)
cat('Error:', round(100*test.error,2),'%')
```

### 4. Augmented self-taught networks 

```r
aug.error <- matrix(NA, nrow=25, ncol=2) # create an empty list to store all models 

# a new list for those model have added feature
new.model.list <- list()

for (i in 1:25){
    k <- 20*i # k
    
    # take the correspond model out from previous training
    model <- model.list[[i]]
    
    # extract features
    feature <- h2o.deepfeatures(model, train, layer=1)
    # combine
    new.train <- h2o.cbind(train, feature) 
    # record the number of features for plotting
    aug.error[i, 1] <- dim(new.train)[2] - 1
    
    # train the model with extra features
    new.model <- h2o.deeplearning(
        x = 2:ncol(new.train), # pixels only
        y = 1, # specify training label
        training_frame = new.train, # specify the frame (imported file)    
        hidden = c(k), # number of layers and their units
        epochs = 50, # maximum number of epoches  
        activation = 'Tanh', # activation function 
        autoencoder = FALSE, # autoencoder is False
    )
    
    # add to the list
    new.model.list[[i]] <- new.model
    
    # calculate the test error
    pred <- h2o.predict(new.model, test)$predict
    aug.error[i, 2] <- error.rate(test$label, pred)
}
```

```r
# plot the error
aug.error.m <- data.frame(f=aug.error[,1], e=aug.error[,2])
# best f and its error
best_f <- data.frame(f=aug.error[which.min(aug.error.m[,2]),1],e=min(aug.error.m[,2]))
# Overall best
f.index <- best_k$k/20
overall_best <- data.frame(f=aug.error[f.index,1],e=aug.error[f.index,2])
# for compare
compare.m <- data.frame(f=aug.error[,1], e=error)
```

```r
ggplot(data=aug.error.m, aes(x=f,y=e)) + ggtitle('The Augmented self-taught networks ') +
    labs(x='Number of features', y='Classification Error') + geom_point(color='blue') + 
    geom_line() + 
    geom_point(aes(x=best_f$f, y=best_f$e), size = 10, colour="red", shape=1) + 
    geom_label(aes(x=1.1*best_f$f, y=best_f$e),
                    label = paste("Best f: ",best_f$f, '\nError: ', round(100 * best_f$e,2), '%')) +
    theme_minimal()
```
