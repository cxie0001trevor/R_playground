# Justify CLT theorem using simulation
Sample a sequence of 10 i.i.d. RVs from a Poisson distribution with $\lambda =10$.
Simulation will take sample size 100, 1000, 10000 and 100000.
Sample mean and standard variance will be computed and final result will be shown in a histogram with theoretical Gaussian curve.

```r
lambda = 10
nRv = 10
f = function(x, mean = cltMeanTheo, sd=cltSdTheo) {
    dnorm(x, mean = mean, sd = sd)
}
printResult = function(size){
    # get sampling ready
    sampleMeans = rep(0,size)
    for (i in 1:size) {
        sampleMeans[i] = mean(rpois(nRv, lambda))
    }
    #Compute results
    cltMeanTheo = lambda
    cltSdTheo = sqrt(lambda/nRv)
    cltMeanEmp = mean(sampleMeans)
    cltsdEmp = sd(sampleMeans)
    #Compare differences
    print(paste("Theoretical Mean:", cltMeanTheo))
    print(paste("Theoretical Sd:", cltSdTheo))
    print(paste("Simulation Mean:", cltMeanEmp))
    print(paste("Simulation Sd:", cltSdTheo))
    #plot histogram and curve
    h = hist(sampleMeans, main = paste(size, " Samples"), breaks = 30, freq = F)
    curve(f, from = min(sampleMeans), to = max(sampleMeans), add = T)
}
```

### Sample size = 100

```r
printResult(100L)
```

### Sample size = 1000

```r
printResult(1000L)
```

### Sample size = 10000

```r
printResult(10000L)
```

### Sample size = 100000

```r
printResult(100000L)
```
