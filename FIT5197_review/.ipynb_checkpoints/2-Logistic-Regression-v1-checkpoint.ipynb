<h1>Logistic Regression</h1>

In this notebook, we start with plotting a simple data set to compare linear and logistic regressions. Then, we will build a number of logistic regression models and discuss their outputs in details.

In this very first example, we use the glass data set that is freely [available](https://archive.ics.uci.edu/ml/machine-learning-databases/glass/glass.data). The dataset contains 214 observations with 9 independent variables and 1 dependent variable (i.e. glass type). For simplicity, we omit the first column (as ID is redundant here) and reduce the domain of the response variable to only two values (to create a binary classification problem). Therefore, the observations will be either window glass (`type=0`) or household glass (`type=1`).

```r
# loading the dataset
## Original Source: https://archive.ics.uci.edu/ml/machine-learning-databases/glass/glass.data
glass <- read.csv('https://archive.ics.uci.edu/ml/machine-learning-databases/glass/glass.data', header = FALSE)
names(glass) = c('ID','Ri','Na','Mg','Al','Si','K','Ca','Ba','Fe','type')
glass$type[glass$type<4] <- 0 # window glasses
glass$type[glass$type>3] <- 1 # other kind of glasses
glass <- glass[ , -1] # drop unneccessary columns
```

# Linear vs. Logistic
OK! Now the dataset is ready. Let's estimate and plot linear regression and logistic regression lines and compare them. Since there are 9 variables and we cannot plot a nine-dimensional plot, we choose only `Al` as the predictor variable and model our response variable `type` accordingly.

Note that we use `glm` function to build a logistic regression. The `glm` parameters are very similar to the parameters of `lm` that we learned previously. However, we have to explicitly mention `family = binomial` to specify that the response variable can only take `0` and `1` values. The `link = logit` is the default value for the binomial response variables, therefore we do not have to explicitly set it.

```r
# Building  models
linear <- lm(type ~ Al, glass)							# train and save a linear regression model
logistic <- glm(type ~ Al, glass, family = binomial)	# train and save a logistic regression model

# Plotting the actual observations
with(glass,						# `with` hellps us to avoid repeating the dataframe name so many times.
  plot(Al,						# the variable in the x axis
       type,					# the variable in the y axis
       pch=20,					# type of the marker
       cex=1.5,					# points size
       col=rgb(0,0,0,0.5),		# 50% transparency to see the points that are co-located better.
       ylim=c(-0.3,1.3),		# the Y axis bounds
       main='Linear vs Logistic'# title of the figure
      ))						# filled circles: the actual class labels (type)

# Plotting the linear reression prediction and line
points(glass$Al,linear$fitted.values, pch=18, col=rgb(0,0,0,0.5))					# filled diamonds: type predicted by linear reression
abline(linear, lty=3, col=rgb(0,0,0,0.5))											# dotted: the linear regression line

#Plotting the logistic regression prediction and curve
points(glass$Al,logistic$fitted.values, pch=1, col=rgb(0,0,0,0.5))					# empty circles: type predicted by logistic regression
curve(predict(logistic,data.frame(Al=x),type="resp"),add=TRUE, col=rgb(0,0,0,0.5))	# solid: the logistic regression curve
```

As the above figure shows, the logistic curve (the solid curve) represents the relationship between these two variables better than the linear regression line (the dashed line). Also note that the values predicted by the logistic regression are bounded by `0` and `1`, so we can consider them as the probability of being from class `1` (i.e. household glass types). In contrast, about 15% of the predictions of the linear regression are out the box bounds.

Although the logistic regression provides us a better prediction, we still cannot directly use the probability provided by the logistic regression as the valid values for `type`. The reason is that `type` only takes `0` and `1`, whilst the probabilities (as the plot shows) are continuous values. 

A common approach to convert the probabilities to acceptable `0` and `1` values is to use a step function. For example, we can conclude that the label (i.e. the value of the response variable) of a sample point is `0` if its likelihood to be a household glass is less than `50%`, and `1` otherwise. For this task, we can simply use a `round` function. 

For a better understanding, we can add a horizontal line that crosses the regression curve at the height `type=0.5` to the above plot. Any point bellow this line would be classified as window glass (i.e. `type=0`), and all the point above this line would be predicted to be household glasses (i.e. `type=1`).

Obviously, the prediction may not always correct. In the bellow figure, we colored the point with green and red colors. The samples which their predicted class matches their real class are colored as green, and the others as red (therefore, the red points indicate the misclassified observations).

```r
x <- (-logistic$coefficients[1])/(logistic$coefficients[2])		# the x value of the vertical line
col = glass$type												# a vector that indicates the color of each point
col[glass$type==round(logistic$fitted.values)] = 'green'		# color the correctprediction as green
col[glass$type!=round(logistic$fitted.values)] = 'red'			# mark the misclassified points with red

with(glass,														# `with` hellps us to avoid repeating the dataframe name 
  plot(Al,														# the variable in the x axis
       type,													# the variable in the y axis
       pch=20,													# type of the marker
       cex=1.5,													# points' size
       col =col,												# points color
       main='True vs False Predictions'							# title of the figure
      ))

curve(predict(logistic,data.frame(Al=x),type="resp"),add=TRUE)	# solid: the logistic regression curve
abline(v=x,lty=2)												# dashed: the vertical line crosses the logistic curve at type=0.5 
abline(h=0.5, lty=3)											# dotted: the horizontal line passes through type=0.5
# For more info refer to http://www.dataschool.io/guide-to-logistic-regression/
```

An interesting observation in the above plot is that the two horizontal and vertical lines divide the points into four categories of True Positive, False Negative, True Negative and False Positive (clockwise starting from the top right). The total numbers of points in these squares are used in many classification evaluation metrics such as accuracy, recall, and precision. Many of these methods are implemented in `Metrics` package.

# Classification Metrics
Let's create a confusion matrix and then calculate a few of the classification evaluation metrics.

```r
confusion.matrix <- as.matrix(table('Actual'=logistic$y, 'Prediction'=round(logistic$fitted.values)))
confusion.matrix
```

As the confusion matrix shows, 157 of window glasses (i.e. `type=0`) and 28 of household glasses (i.e. `type=1`) are predicted correctly. Here, 157 and 28 represent True Negative (TN) and True Positive (TP) values, respectively. If we divide these numbers by the number of samples in each class, then we will have TN and TP rates, respectively.

Our confusion matrix also indicates that 6 window glasses are predicted as the household glass, and 23 household glasses are classified as window glass. 

Let's calculate the accuracy:

```r
N <- nrow(glass)				# number of observations: 214
diag <- diag(confusion.matrix)	# TN and TP: 157 and 6
Accuracy <- sum(diag)/N			# accuracy = (TP + TN)/N = (157 + 28)/214 : 86%
round(Accuracy*100,2)
```

And the distribution of instances over the actual and predicted classes:

```r
rowsums = apply(confusion.matrix, 1, sum)	# number of observations per class
colsums = apply(confusion.matrix, 2, sum)	# number of predictions per class
Actual.Dist = rowsums / N					# distribution of observations over the actual classes
Predicted.Dist= colsums / N					# distribution of observations over the predicted classes
round(data.frame(Actual.Dist,Predicted.Dist)*100,2)
```

Since the `Predicted.Dist` is larger than `Actual.Dist` for `type=0`, we can conclude that the model is more biased towards the class `0`.

Now, let's calculate precision, recall, and f1 metrics for each class

```r
Precision = diag / colsums
Recall = diag / rowsums
F1 = 2 * Precision * Recall / (Precision + Recall) 
round(data.frame(Precision, Recall, F1, Actual.Dist, Predicted.Dist)*100,2) 
# For more implementaions go to: http://blog.revolutionanalytics.com/2016/03/com_class_eval_metrics_r.html
```

# Case Studies
Now we know the basics of the logistic regression. From now on, we analyze the logistic regression outcomes in different scenarios. For all these cases, we use a single [dataset](http://www.ats.ucla.edu/stat/mult_pkg/faq/general/sample.csv) 

```r
dat <- read.csv('http://www.ats.ucla.edu/stat/mult_pkg/faq/general/sample.csv', header = TRUE)
```

```r
dim(dat)
```

As the summary shows, this dataset contains 200 observations of 6 variables. Here, we are interested to study the relationship between `hon` (indicating whether a student is in an honor class) as the response variable and the other 5 variables as the independent variables.

## Zero Independent Variables
A model without any predictor variables is the simplest logistic regression model we can build: 
$t=\beta_0$

```r
model <- glm(hon ~1,data = dat, family = binomial) 
summary(model)
```

The intercept value from the model with no predicor variable is the estimated log odds for the positive class in the whole dataset. In this example, the estimated log odds of being in honors class for the whole population is `-1.1255`, regardless of the student gender or their performance in different topics. 

Recall that we can easily convert the log odds to the probability by the following:

```r
intercept <- model$coefficients[[1]]
exp(intercept)/(1+exp(intercept))
```

Note that `0.245` is the portion of those who took honors versus total number of students. The following shows these portions for both classes:

```r
table(dat$hon)/nrow(dat)
```

## One Binary Independent Variable
Now, we want to build a model that takes gender of the students into account. In this example, we use `female` as a binary variable which `female=1` means the student is a female. Therefore, the model looks like: 
$t=\beta_0+\beta_1 \text{female}$

Before building the model, let's have a quick look at the crosstable and see if gender has any effect in our model. Here, we use `xtab` function to calculate and print hte crosstable.

```r
xtabs(~hon+female, data=dat)
```

This crosstable basically says among al the participates of honors class, 32 of them were female, whilst there are 77 female students that did not take the honors program.

Using this table, we can easily calculate the odds of male and female taking the honors. Then, we can calculate the odds ratio. Let's do it.

```r
(female.odds <- 32/77)
(male.odds <- 17/74)
(female.male.ratio <- female.odds/male.odds)*100
```

The female to male odds ratio is almost $181\%$, which means female have $81\%$ more chance to go through the honors program according to our dataset.

OK! Now let's build the model and interpret the results.

```r
model <- glm(hon ~ female, data = dat, family = binomial) 
summary(model)
```

As the results show, the intercept $-1.47$ which is equvalent to the log odds for males or more formally `log(0.2297) = -1.4709` (recall male is the reference group as `female = 0` means a male).  

The coefficient for female is $0.5928$ which is the log of odds ratio between the females and males or `log(1.8090) = 0.5927`.  This means, we could calculate the odds ratio by exponentiating the coefficient for female or `exp(0.5928)*100`.

## One Continuous Independent Variable
Building a model using a single continuous indipendent variable is also a very easy task. For this example, let's use `math` score as the continuous variable and build: 
$t=\beta_0+\beta_1 \text{math}$

```r
model <- glm(hon ~ math, data = dat, family = binomial) 
summary(model)
```

In this example, the estimated intercept is the log odds of a student being in an honors class with a math score of zero.  So, the chance of being in an honors class when the math score is zero is almost zero (i.e. `exp(-9.7939) = 5.5788e-05`). OK! That is not suprising, especially because no one got that score! Indeed, all math scores are above $30$.

The `math` estimated coefficient is $0.15634$, which means for a one unit increase in the math score, the expected change in log odds is $0.15634$. Since `exp(0.15634)*100 - 100= 16.92`, we can say for every unit increase in math score, we expect to see $16.92\%$ increase in the chance of being in an honors class.

## Multiple Independent Variables
Let's take one further step forward and build a model based on several independent variables (assuming there is no interaction between them). Each of these variables can be discrete or continuous. For example: 
$t=\beta_0+\beta_1 \text{math}+\beta_2 \text{female}+\beta_3 \text{read}$

```r
model <- glm(hon ~ math + female + read, data = dat, family = binomial) 
summary(model)
```

This model says that, holding math and reading scores at a fixed value, 
the chance of getting into an honors for a female is $166.43\%$ higher than the chance of a male if both have the same math and reading score. How we calculate this chance?

Easy! The `female` coefficient is $0.97995$ and `exp(0.97995) * 100 - 100 = 166.43`, which means `female = 1` has $166.43\%$ higher chance of `female = 0` when the other values are hold fixed.

Now, let's tak a look at the `math` coefficient. It says that, for students with the same gender and reading score, every increase in math score will will increase their chance by $13.08\%$ because `exp(0.12296) * 100 - 100 = 13.08`.

## Interaction Terms
Up to this point, we ignored the interaction between the predictor variables. Using interaction terms, we can easily take thoese linkages into account. For example, let's build a model based in `math`, `female` and the interactin between them (`math*female`):
$t=\beta_0+\beta_1 \text{math}+\beta_2 \text{female}+\beta_3 \text{math}\times\text{female}$

```r
model <- glm(hon ~ math + female + math*female, data = dat, family = binomial) 
summary(model)
```

We should note that when an interaction term, we cannot expcet to fix some variables while their interaction term is changing, or vice versa. For example, it is impossible to assume `math` and `math:female` are fixed while `female` is changing.

In this example we have an interaction term of a binary variable (i.e. `female`) and a continuous variable (i.e. `math`). Therefore, we can think that there are two separate equations: 

1. for males (`female=0`): $t=\beta_0+\beta_1 \text{math}$
2. for females (`female=1`): $t=\beta_0+\beta_1 \text{math}+\beta_2 + \beta_3 \text{math} = (\beta_0+\beta_2)+(\beta_1 + \beta_3)\  \text{math}$

Now, we can say that for male students, every unit increase in math score yields a change in log odds of $0.12938$ or $13.81\%$ increase in chance of going to an honors class because `exp(0.12938)*100-100 = 13.81`. Note that, for male students we only look at $\beta_1$.

For the female students, the interpretation is different. For these students, every unit increase in math score yields a change in log odds of $\beta_1 + \beta_3$ which is  `0.12938 + 0.06700 = 0.19638`. This means that every single unit increase in math for female students will result in $21.69\%$ higher chance because `exp(0.12938+0.06700)*100-100 = 21.69`.

The ratio of odds ratios of females over odds ratio of males is `exp(0.12938+0.06700)/exp(0.12938) = 1.0692` which turns out to be the exponentiated coefficient for the interaction term `math:female` or `exp(0.06700) = 1.0692`!
